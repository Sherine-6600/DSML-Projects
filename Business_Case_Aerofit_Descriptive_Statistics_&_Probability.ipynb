**About Aerofit**

Aerofit is a leading brand in the field of fitness equipment. Aerofit provides a product range including machines such as treadmills, exercise bikes, gym equipment, and fitness accessories to cater to the needs of all categories of people.


**Business Problem**

The market research team at AeroFit wants to identify the characteristics of the target audience for each type of treadmill offered by the company, to provide a better recommendation of the treadmills to the new customers. The team decides to investigate whether there are differences across the product with respect to customer characteristics.

The dataset has the following features:

***Product***:	KP281, KP481, or KP781

***Age***:	In years

***Gender***:	Male/Female

***Education***:	In years

***MaritalStatus***:	Single or partnered

***Usage***:	The average number of times the customer plans to use the treadmill each week.

***Income***:	Annual income (in $)

***Fitness***:	Self-rated fitness on a 1-to-5 scale, where 1 is the poor shape and 5 is the excellent shape.

***Miles***:	The average number of miles the customer expects to walk/run each week

```python
df1 = pd.crosstab(df['Gender'],
            df['Product'],
            margins=True,
            margins_name='Total').reset_index()
df1
```

```python
print('Probability of Male customer:',np.round(df1.iloc[1,4]/df1.iloc[2,4],2))
print('Probability of Female customer:',np.round(df1.iloc[0,4]/df1.iloc[2,4],2))
print('Probability of given Male customer using KP281:',np.round(df1.iloc[1,1]/df1.iloc[1,4],2))
print('Probability of given Male customer using KP481:',np.round(df1.iloc[1,2]/df1.iloc[1,4],2))
print('Probability of given Male customer using KP781:',np.round(df1.iloc[1,3]/df1.iloc[1,4],2))
print('Probability of given Female customer using KP281:',np.round(df1.iloc[0,1]/df1.iloc[0,4],2))
print('Probability of given Female customer using KP481:',np.round(df1.iloc[0,2]/df1.iloc[0,4],2))
print('Probability of given Female customer using KP781:',np.round(df1.iloc[0,3]/df1.iloc[0,4],2))
```

```python
df2 = pd.crosstab(df['Gender'],
            df['Fitness'],
            margins=True,
            margins_name='Total').reset_index()
df2
```

```python
print('Probability of given Female customer is Fit:',np.round(df2.iloc[0,4]/df2.iloc[0,6],2)+np.round(df2.iloc[0,5]/df2.iloc[0,6],2))
print('Probability of given Female customer is Unfit:',np.round(df2.iloc[0,1]/df2.iloc[0,6],2)+np.round(df2.iloc[0,2]/df2.iloc[0,6],2)+np.round(df2.iloc[0,3]/df2.iloc[0,6],2))
print('Probability of given Male customer is Fit:',np.round(df2.iloc[1,4]/df2.iloc[1,6],2)+np.round(df2.iloc[1,5]/df2.iloc[1,6],2))
print('Probability of given Male customer is Unfit:',np.round(df2.iloc[1,1]/df2.iloc[1,6],2)+np.round(df2.iloc[1,2]/df2.iloc[1,6],2)+np.round(df2.iloc[1,3]/df2.iloc[1,6],2))
```

```python
df3 = pd.crosstab(df['MaritalStatus'],
            df['Product'],
            margins=True,
            margins_name='Total').reset_index()
df3
```

```python
print('Probability of Single customer:',np.round(df3.iloc[1,4]/df3.iloc[2,4],2))
print('Probability of Partnered customer:',np.round(df3.iloc[0,4]/df3.iloc[2,4],2))
print('Probability of given Single customer using KP281:',np.round(df3.iloc[1,1]/df3.iloc[1,4],2))
print('Probability of given Single customer using KP481:',np.round(df3.iloc[1,2]/df3.iloc[1,4],2))
print('Probability of given Single customer using KP781:',np.round(df3.iloc[1,3]/df3.iloc[1,4],2))
print('Probability of given Partnered customer using KP281:',np.round(df3.iloc[0,1]/df3.iloc[0,4],2))
print('Probability of given Partnered customer using KP481:',np.round(df3.iloc[0,2]/df3.iloc[0,4],2))
print('Probability of given Partnered customer using KP781:',np.round(df3.iloc[0,3]/df3.iloc[0,4],2))
```

--> Model KP281 & KP481 are more popular amoong customers in income slab between USD 35,000 to USD 60,000. They should be targeted by Aerofit to sell them these models. Some offers/discounts can also be provided the company to attract these customers.
